```python
import numpy as np
import qiskit
from qiskit import visualization, QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute
%matplotlib inline
from math import pi
from Circuits import getBloch, stateGenerator, getMeasure, makeCirc
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
circ = makeCirc(6, 0)
getBloch(circ, backend)
```

```python
circ.cswap(1, 2, 3)
circ.cswap(1, 4, 5)
circ.cswap(0, 3, 5)

for i in range(2, 6):
    circ.measure(i, i)
```

```python
circ.draw('mpl')
```

```python
getMeasure(circ, 1, backend)[0][::-1]
```

```python
circ = QuantumCircuit(6)
circ.cswap(1, 2, 3)
circ.cswap(1, 4, 5)
circ.cswap(0, 3, 5)
```

```python
mux = circ.to_gate(label=' Mux')
```

```python
circ2 = QuantumCircuit(12)
circ2.append(mux, [0, 1, 2, 3, 4, 5])
```

```python
circ2.draw('mpl')
```

```python
circ3 = QuantumCircuit(6)
circ3.cswap(1, 2, 3)
circ3.cswap(0, 2, 4)
circ3.cswap(0, 3, 5)
```

```python
circ3.draw('mpl')
```

```python
demux = circ3.to_gate(label=' Demux')
```

```python
circ2.append(demux, [6, 7, 8, 9, 10, 11])
```

```python
circ2.draw('mpl')
```
