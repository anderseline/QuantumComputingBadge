# Bloch Sphere

```python
from qiskit import visualization, QuantumCircuit, Aer, execute, assemble
%matplotlib inline
from math import pi, sqrt
import pylatexenc
```

```python
circ = QuantumCircuit(6)
```

```python
circ.h(1)
circ.x(2)
circ.y(3)
circ.z(4)
circ.h(5)
circ.rz(pi/4, 5)
```

```python
circ.draw('mpl')
```

```python
qobj = assemble(circ)
backend = Aer.get_backend('statevector_simulator')
state = backend.run(qobj).result().get_statevector()
```

```python
visualization.plot_bloch_multivector(state)
```

Shows how each of the qubits changed from their initial state (q0)
 - Hadamard gate was applied to q1
 - 180 degree rotation about x applied to q2
 - 180 degree rotation about y applied to q3 (appears the same as q2)
 - 180 degree rotation about z applied to q4 (doesn't change direction)
 - Hadamard gate, then 45 degree rotation about z applied to q5
