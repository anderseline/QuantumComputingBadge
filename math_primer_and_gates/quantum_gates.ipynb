# Gates, Pauli Matrices, and Phase Shift

```python
import numpy as np 
from qiskit import visualization, QuantumCircuit, Aer, execute
%matplotlib inline
from math import pi, sqrt
import pylatexenc
```

```python
# initialize 5 empty qubits 
circ = QuantumCircuit(2)
```

```python
circ.h(0) #hadamard gate
circ.x(1) #pauli x
```

```python
# shows the logic flow of each qubit thorugh the gates 
circ.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator') # initialize backend simulation
result = execute(circ, backend).result() # run simulation 
output = result.get_statevector(circ, decimals=3) # get result 
```

```python
visualization.plot_state_city(output) # plots all possible values of qubits 0-4
```

```python
# now with using 5 qubits 
circ = QuantumCircuit(5)

circ.h(0) #hadamard gate
circ.x(1) #pauli x
circ.y(2) #pauli y
circ.z(3) #pauli z
circ.rz(pi/4, 4) #phase shift (?)
circ.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator') # initialize backend simulation
result = execute(circ, backend).result() # run simulation 
output = result.get_statevector(circ, decimals=3) # get result 
```

```python
visualization.plot_state_city(output)
```
