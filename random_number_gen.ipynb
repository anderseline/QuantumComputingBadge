# Random Number Generator from 1-64

```python
from qiskit import visualization, QuantumCircuit, Aer, execute, assemble
%matplotlib inline
from math import pi, sqrt
import pylatexenc
```

```python
circ = QuantumCircuit(6, 6) # 6 quantum, 6 classical bits
```

```python
for i in range(0, 6):
    circ.h(i)
    circ.measure(i, i)
```

```python
circ.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
# generates 10 random numbers between 1-64
measures = []
for j in range(0, 10):
    outcome = execute(circ, backend).result()
    hist = outcome.get_counts()
    for i in hist.keys():
        measures.append(int(i, 2))
```
