# Fredkin Gate
<i>CSWAP Gate</i>

```python
import numpy as np
from qiskit import visualization, QuantumCircuit, Aer, execute, assemble, QuantumRegister, ClassicalRegister
%matplotlib inline
from math import pi, sqrt
import pylatexenc
```

```python
qreg = QuantumRegister(3)
qclas = ClassicalRegister(3)
quantumC = QuantumCircuit(qreg, qclas)
initializedState = [0 for i in range(0, 2**3)]
initializedState[5] = 1
quantumC.initialize(initializedState, qreg)
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
visualization.plot_bloch_multivector(execute(quantumC, backend).result().get_statevector())
```

```python
quantumC.cswap(0, 1, 2) #create Fredkin gate
```

```python
quantumC.draw('mpl')
```

Qubit 0 is the control, with 1 and 2 being swapped

```python
visualization.plot_bloch_multivector(execute(quantumC, backend).result().get_statevector())
```

Qubits 1 and 2 are swapped since qubit 0 is in state 1

```python
# create cswap gate using Toffoli and cnot gates

quantumC.cx(1, 2,) #CNOT
quantumC.ccx(0, 2, 1) #Toffoli
quantumC.cx(1, 2) #CNOT

quantumC.draw('mpl')
```

This circuit will actually swap qubits 1 and 2 twice

```python
visualization.plot_bloch_multivector(execute(quantumC, backend).result().get_statevector())
```
