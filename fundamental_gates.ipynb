# Fundamental Quantum Gates

Toffoli (AND), Swap, Phase Shift

```python
import numpy as np
from qiskit import visualization, QuantumCircuit, Aer, execute, assemble, QuantumRegister, ClassicalRegister
%matplotlib inline
from math import pi, sqrt
import pylatexenc
```

```python
n = 7
qreg = QuantumRegister(n)
qclas = ClassicalRegister(n)
quantumC = QuantumCircuit(qreg, qclas)
initializedState = [0 for i in range(0, 2**n)]
initializedState[107] = 1
quantumC.initialize(initializedState, qreg)
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
visualization.plot_bloch_multivector(execute(quantumC, backend).result().get_statevector())
```

```python
quantumC.ccx(0, 1, 2) # toffoli
quantumC.swap(3, 4) #swap
quantumC.h(6) #hadamard
quantumC.cp(pi/3, 5, 6) #phase shift
```

```python
visualization.plot_bloch_multivector(execute(quantumC, backend).result().get_statevector())
```

- Toffoli: Since qubit 0 and qubit 1 are both 1 cat vectors, qubit 2 switches from a 0 cat to a 1 cat vector 
- Swap: Qubits 3 and 4 have changed state (0 / 1 cat)
- Hadamard: Qubit 6 undergoes a 135 degree rotation, and a reflection across y = x (points 45 degrees above x/y plane in 0 vector direction)
- Phase Shift: Since qubit 5 is 1, qubit 6 undergoes a further phase shift of 60 degrees towards the x direction
